```python
from math import sin
def newton(fn,dfn,x,tol,maxiter):
    for i in range(maxiter):
        xnew = x - fn(x)/dfn(x)
        if abs(xnew-x)<tol:                          
            break
        x = xnew
    return xnew, i

y = lambda x: 2*x**2 - 5*x + 3 
dy = lambda x : 4*x - 5 

x, n = newton(y, dy, 1.4, 0.0001, 100)
print('the root is %.3f at %d iterations.'%(x,n))
```

```python
from math import sin
def secant(fn,x1,x2,tol,maxiter):
    for i in range(maxiter):
        xnew  = x2 - (x2-x1)/(fn(x2)-fn(x1))*fn(x2)
        if abs(xnew-x2) < tol:
            break
        else:
            x1 = x2
            x2 = xnew
    else:
        print('warning: Maximum number of iterations is reached')
    return xnew, i

f = lambda x: 2*x**2 - 5*x + 3 

x1 = float(input('enter x1: '))
x2 = float(input('enter x2: '))

r, n = secant(f,x1,x2,1.0e-6,100)

print('Root = %f at %d iterations'%(r,n))
```

```python
import numpy as np
import matplotlib.pyplot as plt
f=lambda x: 0.1*x**5 - 0.2*x**3 + 0.1*x - 0.2

h = 0.01
#will create an array of elements between -1 and 1 having 45 equal
x = np.linspace(-1,1,80)
# forward difference

dff1 = (f(x+1)-f(x))/h
dff2 = (f(x+2)-2*f(x+1)+f(x))/h**2
#plot
plt.plot(x,f(x),'-k',x,dff1,'-g',x,dff2,'-r')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(["f(x)","f'(x)","f''(x)"])
plt.grid()
```
